```python
def calculate_delta_teta(s,pre,Label,x1,x2,x3,x4,i):

    dbias = 2*(s-Label[i])*(1-s)*s
    dteta1 = X1[i]*dbias
    dteta2 = X2[i]*dbias
    dteta3 = X3[i]*dbias
    dteta4 = X4[i]*dbias

    return dteta1,dteta2,dteta3,dteta4,dbias
```

```python
def update_teta(dteta1,dteta2,dteta3,dteta4,dbias,teta1,teta2,teta3,teta4,bias,alpha):

    teta1 = teta1-dteta1*alpha
    teta2 = teta2-dteta2*alpha
    teta3 = teta3-dteta3*alpha
    teta4 = teta4-dteta4*alpha
    bias = bias-dbias*alpha

    return teta1,teta2,teta3,teta4,bias
```

```python
def calculate_error(Label,s,i,error_array):

    error = (Label[i]-s)**2
    error_array[i]=error

    return error
```

```python
def average_error_per_epoch(average_error_array,error_array):

    avg_error = sum(error_array) / float(len(error_array))
    average_error_array.append(avg_error)

    return avg_error
```

```python
def predict(X1,X2,X3,X4,teta1,teta2,teta3,teta4,bias):

    h = X1*teta1+X2*teta2+X3*teta3+X4*teta4+bias
    s = s(h)

    if s<0.5:
        pred = 0
        notes = "Iris-setosa"
    else:
        pred = 1
        notes = "Iris-versicolor"
    
    return pred, notes, s
```

```python
def validation(vX1,vX2,vX3,vX4,Label_val,teta1,teta2,teta3,teta4,bias,validation_array):

    s_array = []
    tmp = []

    for i in range(len(vX1)):
        _, _, s = predict(vX1[i],vX2[i],vX3[i],vX4[i],teta1,teta2,teta3,teta4,bias)
        s_array.append(s)

    for i in range(len(s_array)):
        error = (Label_val[i]-s_array[i])**2
        tmp.append(error)

    average_error_per_epoch(validation_array,tmp)

    return 0
```

```python
def plot(validation_array,average_error_array):

    fig = plotting.figure()
    fig.suptitle('MSE Graph')
    seaborn.set_style("white")

    plotting.plot(list(range(1,61)),validation_array, Label = "Validation")
    plotting.plot(list(range(1,61)),average_error_array, Label = "Training")

    plotting.xLabel('Epoch')
    plotting.yLabel('MSE')
    plotting.legend()

    return plotting.show()
```
